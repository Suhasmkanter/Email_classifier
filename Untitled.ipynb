```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.model_selection import cross_val_score 
from sklearn.model_selection import train_test_split,cross_val_score ,GridSearchCV
from sklearn.linear_model import LogisticRegression 
from sklearn.ensemble import RandomForestClassifier
# fetch dataset 

spambase = fetch_ucirepo(id=94) 
  
# data (as pandas dataframes) 
X = spambase.data.features 
y = spambase.data.targets 
  
data = pd.concat([X,y],axis=1)
data.head()
```

```python
from wordcloud import WordCloud 

value1  = X[y['Class'] == 1]
value2 = X [y ['Class'] == 0 ]
spam = {}
for i in value1.columns: 
    if i != 'Class': 
      spam[i] = value1[i].mean()
not_spam = {}
for i in value2.columns: 
    if i != 'Class': 
      not_spam[i] = value2[i].mean()

spam
spambro = WordCloud(width=1000,height=900,background_color='black').generate_from_frequencies(spam)
notspambro = WordCloud(width=1000,height=900,background_color='black').generate_from_frequencies(not_spam)
print(spambro,notspambro) 
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title("Spam words")
plt.imshow(spambro, interpolation='bilinear')

plt.subplot(122)
plt.title("Not Spam Words")
plt.imshow(notspambro, interpolation='bilinear')

plt.show()

```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
scaler.fit(data[['capital_run_length_longest','capital_run_length_total']])
scaled_data = scaler.transform(data[['capital_run_length_longest','capital_run_length_total']])
print(scaled_data)
maindatabro = pd.concat([new_data,pd.DataFrame(scaled_data,columns=['capital_run_length_longest','capital_run_length_total'])],axis = 1)
maindatabro.head(5)
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split 
model = LogisticRegression(max_iter = 5000)
X_train,X_test,y_train,y_test = train_test_split(new_data.drop('Class',axis=1 ),new_data['Class'],test_size=0.2,random_state = 42)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
from sklearn.metrics import accuracy_score 
print("The Accuracy Score : ")
print( accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

y_prob = model.predict_proba(X_test)[:,1]

auc = roc_auc_score(y_test, y_prob)
print("ROC AUC Score:", auc)

# Plotting ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()



values = cross_val_score(LogisticRegression(max_iter=5000),new_data.drop('Class',axis=1 ),new_data['Class'],cv=10).mean()
print(values)
```

```python
params_grid={
    "max_iter":[1000,2000,3000,4000,5000]
}
modelfit = GridSearchCV(LogisticRegression(),params_grid,cv=5)
print(modelfit)
modelfit.fit(X_train, y_train)

# To see the best model and best parameters
print("Best Parameters:", modelfit.best_params_)
print("Best Estimator:", modelfit.best_estimator_)

# Predict with best model
y_pred = modelfit.predict(X_test)

# Then evaluate (accuracy, confusion matrix, etc.)
from sklearn.metrics import accuracy_score
print("Accuracy after GridSearch:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Get predictions
y_pred = model.predict(X_test)

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Spam", "Spam"])
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
params_grids = {
    "n_estimators":[100,200,300],
    "max_depth":[10,20,30]
}
rfe = GridSearchCV(RandomForestClassifier(random_state=42,criterion='gini'),params_grids,cv=5)
rfe.fit(X_train,y_train)
y_prediction  = rfe.predict(X_test)
print("The accuracy score is : ")
print(accuracy_score(y_test,y_prediction))
```

```python
importantFeatures = rfe.best_estimator_.feature_importances_
features = [i for i in new_data.columns if i != 'Class']
print(len(importantFeatures),len(features))

df = pd.DataFrame({'feature':features,'FeatureImportance':importantFeatures})
valuesdf = df.sort_values(by='FeatureImportance',ascending=False)
```
